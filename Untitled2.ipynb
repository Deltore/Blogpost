```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
url_aantal = 'https://data.rivm.nl/covid-19/COVID-19_aantallen_gemeente_per_dag.json'
url_ziekenhuis = 'https://data.rivm.nl/covid-19/COVID-19_ziekenhuisopnames.json'
url_testen = 'https://data.rivm.nl/covid-19/COVID-19_uitgevoerde_testen.json'

r_aantal = requests.get(url_aantal)
json_data_aantal = r_aantal.json()

r_ziekenhuis = requests.get(url_ziekenhuis)
json_data_ziekenhuis = r_ziekenhuis.json()

r_testen = requests.get(url_testen)
json_data_testen = r_testen.json()

df_aantal = pd.DataFrame(json_data_aantal)
df_ziekenhuis = pd.DataFrame(json_data_ziekenhuis)
df_testen = pd.DataFrame(json_data_testen)
```

```python
#We groupen de dataframe van de ziekenhuisopnames op de dagen, 
#zodat we het aantal ziekenhuisopnames hebben per dag 
df_aantal_nieuw = pd.DataFrame(df_aantal.groupby('Date_of_publication')['Hospital_admission'].sum())
df_aantal_nieuw['Date_of_publication'] = df_aantal_nieuw.index

df_ziekenhuis_nieuw = pd.DataFrame(df_ziekenhuis.groupby('Date_of_statistics')['Hospital_admission'].sum())
df_ziekenhuis_nieuw['Date_of_statistics'] = df_ziekenhuis_nieuw.index
```

```python
#Deze cel maakt een plot met een slider. De slider geeft het aantal dagen aan sinds het begin van de 
#pandemie

day = widgets.IntSlider(
    value = 0,
    min = 0,
    max = 369,
    step = 1,
    description = 'Day',
    orientation = 'horizontal',
    layout=widgets.Layout(width='50%'))

def get_count(day):
    admissions = df_ziekenhuis_nieuw['Hospital_admission'][day]
    date = df_ziekenhuis_nieuw['Date_of_statistics'][day]
    fig = plt.figure()
    
    s = fig.add_subplot()
    s.bar([1], [admissions], width=1, tick_label=[date])
    plt.ylim(0, 695)
    plt.ylabel('# Ziekenhuisopnames')
    plt.title ('Aantal ziekenhuisopnames per dag')
    
out = widgets.interactive_output(get_count, {'day': day})
widgets.VBox([day, out])
```

```python
#Aantal ziekenhuisopnames per veiligheidsregio
df3 = pd.DataFrame(df_aantal.groupby('Security_region_name')['Hospital_admission'].sum())

#Aantal testen per veiligheidsregio
df4 = pd.DataFrame(df_testen.groupby('Security_region_name')['Tested_with_result'].sum())

#Aantal positief geteste mensen per veiligheidsregio
df_positive = pd.DataFrame(df_testen.groupby('Security_region_name')['Tested_positive'].sum())

#verwijder de rij met onbekende waarde
df_positive = df_positive.drop('Onbekend')

#Voeg de tabel met het aantal ziekenhuisopnames en de tabel met positieve testen samen
df_join = df3.merge(df_positive, on='Security_region_name')


df_join
```

```python
#Deze cel maakt een scatterplot aan met op de x-as het aantal ziekenhuisopnames 
#en op de y-as het aantal positieve testen. De checkbox bepaald of de regressielijn wordt getoont.

a = widgets.Checkbox(
    value = True,
    description = 'Show regression',
    disabled = False)

sns.set_style('whitegrid')
sns.set_context('talk')

def plot(a):
    if a == True:
        sns.lmplot(x='Hospital_admission', y='Tested_positive', data=df_join, height=8, ci=None)
        plt.xlim(0, 3000)
        plt.ylim(0, 90000)
        plt.xlabel('Aantal ziekenhuisopnames')
        plt.ylabel('Aantal positieve testen')
        plt.title('spreidingsdiagram ziekenhuisopnames t.o.v positieve testen')
    else:
        sns.relplot(kind = 'scatter', x='Hospital_admission', y='Tested_positive', data=df_join, height=8)
        plt.xlim(0, 3000)
        plt.ylim(0, 90000)
        plt.xlabel('Aantal ziekenhuisopnames')
        plt.ylabel('Aantal positieve testen')
        plt.title('spreidingsdiagram ziekenhuisopnames t.o.v positieve testen')
        
out = widgets.interactive_output(plot, {'a': a})
widgets.VBox([a, out])
```

```python
provincies = ['Noord-Holland', 'Noord-Brabant', 'Noord-Brabant', 'Drenthe', 'Flevoland', 'Friesland', 'Gelderland',
 'Gelderland', 'Noord-Holland', 'Groningen', 'Zuid-Holland', 'Zuid-Holland', 'Zuid-Holland',
 'Noord-Holland', 'Limburg', 'Limburg', 'Noord-Brabant', 'Gelderland', 'Noord-Holland', 
 'Zuid-Holland', 'Overijssel', 'Utrecht', 'Noord-Holland', 'Zeeland', 'Zuid-Holland']

inwoners = [493657, 422979, 649944, 2086142, 586061, 1117164, 2563432, 2879611, 1162500, 1354979, 383519, 3708585]
```

```python
#Relatief aantal geteste mensen per provincie

df4 = df4.drop('Onbekend')

#we voegen de provincie toe als kolom aan de dataframe per veiligheidsregio
df4['provincies'] = provincies
#We voegen de veiligheidsregios samen zodat we data hebben per provincie
df5 = pd.DataFrame(df4.groupby('provincies')['Tested_with_result'].sum())

df5['provincies'] = df5.index
df5['Inwoners'] = inwoners

#deel het aantal geteste mensen door het totaal aantal inwoners per provincie
df5['Getest_relatief'] = df5['Tested_with_result']/df5['Inwoners']
df5
```

```python
#we voegen de provincie toe als kolom aan de dataframe per veiligheidsregio
df3['provincies'] = provincies

#We voegen de veiligheidsregios samen zodat we data hebben per provincie
df6 = pd.DataFrame(df3.groupby('provincies')['Hospital_admission'].sum())

df6['provincies'] = df6.index
df6['Inwoners'] = inwoners

#deel het aantal ziekenhuisopnames door het totaal aantal inwoners per provincie
df6['ziekenhuis_opnames_relatief'] = df6['Hospital_admission']/df6['Inwoners']
df6
```

```python
df6['Getest_relatief']=df5['Getest_relatief']
df6['Tested_with_result']=df5['Tested_with_result']
df6
```

```python
#Deze cel maakt een drop down menu waar je een provincie kan kiezen. 
#Vervolgens krijg je informatie over de provincie

prov = widgets.Dropdown(
    options = df6['provincies'])


def info(prov):
    hosp = df6.loc[prov]['Hospital_admission']
    test = df6.loc[prov]['Tested_with_result']
    print('Het aantal ziekenhuisopnames in ' + str(prov) + ' is: ' + str(hosp))
    print('Het aantal geteste mensen in ' + str(prov) + ' is: ' + str(test))

out = widgets.interactive_output(info, {'prov': prov})
widgets.VBox([prov, out])
```
